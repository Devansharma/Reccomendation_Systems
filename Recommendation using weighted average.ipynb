#  Recommendation using Weighted Average

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
credits = pd.read_csv(r"C:\Users\Mahesh Sharma\Downloads\tmdb_5000_credits.csv")
```

```python
movies = pd.read_csv(r'C:\Users\Mahesh Sharma\Downloads\tmdb_5000_movies.csv')
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})
movies_df = movies.merge(credits_column_renamed, on='id')
movies_df.head()
```

```python
movie_cleaned = movies_df.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
```

```python
movie_cleaned.info()
```

```python
v= movie_cleaned['vote_count']
R = movie_cleaned['vote_average']
C = movie_cleaned['vote_average'].mean()
m = movie_cleaned['vote_count'].quantile(0.70)
```

```python
movie_cleaned['weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
movie_cleaned.head()
```

```python
movie_sorted_ranking = movie_cleaned.sort_values('weighted_average', ascending=False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
scalar = MinMaxScaler()
movie_scaled = scalar.fit_transform(movie_cleaned[['weighted_average', 'popularity']])
movie_normalized = pd.DataFrame(movie_scaled, columns=['weighted_average', 'popularity'])
movie_normalized.head()
```

```python
movie_cleaned[['normalized_weight_average', 'normalized_popularity']]=movie_normalized
```

```python
movie_cleaned.head()
```

```python
movie_cleaned['score'] = movie_cleaned['normalized_weight_average']*0.5+movie_cleaned['normalized_popularity']*0.5
movie_scored = movie_cleaned.sort_values(['score'], ascending=False)
movie_scored[['original_title', 'normalized_weight_average', 'normalized_popularity','score']].head(20)
```
